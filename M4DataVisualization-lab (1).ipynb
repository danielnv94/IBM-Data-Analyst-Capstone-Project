# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
from matplotlib import pyplot as plt
from matplotlib.ticker import ScalarFormatter, PercentFormatter
import seaborn as sns
```

```python
QUERY = """
SELECT ConvertedComp
FROM master
order by ConvertedComp
"""
df = pd.read_sql_query(QUERY,conn)

#Create figure and axes
fig, ax = plt.subplots(figsize=(8, 4))
count = ax.hist(df)

for rect in ax.patches:
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width() / 2, height, f'{int(height)}', ha='center', va='bottom')

ax.set_title('Annual Compensation in USD') 
ax.set_xlabel('Salary in USD')
ax.set_ylabel('Number of  Respondents')
plt.gca().xaxis.set_major_formatter(ScalarFormatter(useMathText=True))
plt.ticklabel_format(style='plain', axis='x')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
QUERY = """
SELECT Age, WorkWeekHrs
FROM master
"""

df_age_workWeekHrs = pd.read_sql_query(QUERY,conn)

df_age_workWeekHrs.plot(kind='scatter', x='Age', y='WorkWeekHrs', figsize=(10, 6), color='darkblue')
plt.title('Hours Worked per Week by Age')
plt.xlabel('Age')
plt.ylabel('Hours Worked per Week')
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
conn = sqlite3.connect("m4_survey_data.sqlite")

QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
"""

df_age_workWeekHrs1 = pd.read_sql_query(QUERY,conn)

size=df_age_workWeekHrs1['Age']
fig=plt.figure(figsize=(14, 7))
sns.scatterplot(data=df_age_workWeekHrs1, x='CodeRevHrs', y='WorkWeekHrs',hue="Age", size=size, legend='brief')
plt.title('Correlation Between Hrs Worked/Week and Hrs Reviewing Code by Age')
plt.xlabel('Hours Reviweing Code')
plt.ylabel('Hours Worked per Week')
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
table_name = 'DatabaseDesireNextYear'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

```python
QUERY = """
SELECT DatabaseDesireNextYear, COUNT (*) as Count
FROM DatabaseDesireNextYear
group by DatabaseDesireNextYear
order by Count DESC
"""

df_DatabaseDesireNextYear = pd.read_sql_query(QUERY,conn)
                                              
top_databases = df_DatabaseDesireNextYear['DatabaseDesireNextYear'].head(5)
percentages = df_DatabaseDesireNextYear['Count'].head(5)

fig,ax=plt.subplots()
ax.pie(percentages, labels = top_databases, colors = ['gold','blue','lightgreen','coral','cyan'], autopct='%1.1f%%',explode = [0,0,0,0,0.1])
ax.set_aspect('equal')
plt.title('Top 5 Databases Respondents Want to Learn for Next Year')
plt.show()
```

```python
QUERY = """
SELECT LanguageDesireNextYear, COUNT (*) as Count
FROM LanguageDesireNextYear
group by LanguageDesireNextYear
order by Count DESC 
"""

df_LanguageDesireNextYear = pd.read_sql_query(QUERY,conn)
df_LanguageDesireNextYear.head()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs
FROM master
WHERE Age BETWEEN 30 AND 35
"""

df_age_workWeekHrs2 = pd.read_sql_query(QUERY,conn)

median_WorkWeekHrs = df_age_workWeekHrs2["WorkWeekHrs"].median()
median_CodeRevHrs = df_age_workWeekHrs2["CodeRevHrs"].median()

plt.bar('Metrics', median_CodeRevHrs, label='Median WorkWeekHrs', color='green')
plt.bar('Metrics', median_WorkWeekHrs, bottom=median_CodeRevHrs, label='Median CodeRevHrs', color='blue')

plt.xlabel('Metrics')
plt.ylabel('Median Hours')
plt.title('Median WorkWeekHrs and CodeRevHrs')
plt.legend()
plt.show()

print("Median WorkWeekHrs:", median_WorkWeekHrs)
print("Median CodeRevHrs:", median_CodeRevHrs)
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
QUERY = """
SELECT ConvertedComp, Age
FROM master
WHERE Age BETWEEN 45 AND 60
GROUP BY Age
"""

df_ConvertedComp = pd.read_sql_query(QUERY,conn)

ConvertedComp_median = []

for age_group, group_data in df_ConvertedComp.groupby('Age'):
    median_converted_comp = group_data['ConvertedComp'].median()
    ConvertedComp_median.append((age_group, median_converted_comp))

median_comp_df = pd.DataFrame(ConvertedComp_median, columns=['Age', 'Median_ConvertedComp'])
print(median_comp_df)

fig, ax = plt.subplots()

ax.plot(median_comp_df['Age'], median_comp_df['Median_ConvertedComp'], marker='s', markersize=5, color='green', linestyle="dotted")

plt.grid(True)
ax.set_title('Median Annual Compensation by Age Group') 
ax.set_xlabel('Age')
ax.set_ylabel('Median Annual Compensation (Thousand USD)')
plt.show()
```

### Bar Chart


```python
QUERY = """
SELECT ConvertedComp, Age
FROM master
WHERE Age BETWEEN 25 AND 30
GROUP BY Age
"""

df_ConvertedComp = pd.read_sql_query(QUERY,conn)

ConvertedComp_median = []

for age_group, group_data in df_ConvertedComp.groupby('Age'):
    median_converted_comp = group_data['ConvertedComp'].median()
    ConvertedComp_median.append((age_group, median_converted_comp))

median_comp_df = pd.DataFrame(ConvertedComp_median, columns=['Age', 'Median_ConvertedComp'])
print(median_comp_df)

fig, ax = plt.subplots()

ax.plot(median_comp_df['Age'], median_comp_df['Median_ConvertedComp'], marker='s', markersize=5, color='green', linestyle="dotted")

plt.grid(True)
ax.set_title('Median Annual Compensation by Age Group') 
ax.set_xlabel('Age')
ax.set_ylabel('Median Annual Compensation (Thousand USD)')
plt.show()
```
